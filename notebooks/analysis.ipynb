## 1. Habits vs performance (D2, n=1,000)
What actually drives exam scores?

```python
import pandas as pd, numpy as np
d2 = pd.read_csv('../data/raw/d2_habits_performance.csv')
corr = d2.select_dtypes('number').corr()['exam_score'].drop('exam_score').sort_values(key=abs, ascending=False)
corr.round(3)
```

Study time dominates (**r ≈ 0.83**); screen time is weakly negative.

## 2. The mental-health gap (D1, n=101)

```python
d1 = pd.read_csv('../data/raw/d1_mental_health_survey.csv')
d1.columns = ['ts','gender','age','course','year','cgpa','marital','depression','anxiety','panic','treatment']
yn = {'Yes':1,'No':0}
for c in ['depression','anxiety','panic','treatment']: d1[c]=d1[c].map(yn)
print('depression %:', round(d1.depression.mean()*100,1))
print('any distress %:', round((d1[['depression','anxiety','panic']].sum(1)>0).mean()*100,1))
print('sought treatment %:', round(d1.treatment.mean()*100,1))
```

## 3. Income by education (D3, n=2,000)

```python
d3 = pd.read_csv('../data/raw/d3_socioeconomic.csv')
d3.groupby('Education')['Income'].mean().round(0).sort_values(ascending=False)
```
